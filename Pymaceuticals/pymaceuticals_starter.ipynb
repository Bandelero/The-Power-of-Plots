## Observations and Insights 

```python
# Dependencies and Setup
import pandas as pd
import random
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import sem
import os
from sklearn import datasets
from scipy.stats import linregress
import sklearn.datasets as dta
import scipy.stats as st
from scipy import stats

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_df = pd.merge(mouse_metadata, study_results, how='outer', on='Mouse ID')
# Display the data table for preview
combined_df.head()
```

## Summary Statistics

```python

#Use groupby to create a new dataframe and calculate mean, median, mode separately.

# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

#Groupbys
sum_stats_df_mean = cleaned_combined_df.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].mean()
sum_stats_df_median = cleaned_combined_df.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].median()
sum_stats_df_variance = cleaned_combined_df.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].var()
sum_stats_df_std = cleaned_combined_df.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].std()
sum_stats_df_sem = cleaned_combined_df.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].sem()

#Merges
sum_stats_merge = pd.merge(sum_stats_df_mean, sum_stats_df_median, how='left', on=['Drug Regimen'])
sum_stats_merge.rename(columns = {'Tumor Volume (mm3)_x': 'Tumor Volume (mm3): mean',
                                  'Tumor Volume (mm3)_y': 'median'}, inplace=True)
sum_stats_merge2 = pd.merge(sum_stats_merge, sum_stats_df_variance, how='left', on=['Drug Regimen'])
sum_stats_merge3 = pd.merge(sum_stats_merge2, sum_stats_df_std, how='left', on=['Drug Regimen'])
sum_stats_merge3.rename(columns = {'Tumor Volume (mm3)_x': 'variance',
                                  'Tumor Volume (mm3)_y': 'std'}, inplace=True)
sum_stats_merge4 = pd.merge(sum_stats_merge3, sum_stats_df_sem, how='left', on=['Drug Regimen'])
sum_stats_merge4.rename(columns = {'Tumor Volume (mm3)': 'sem'}, inplace=True)
sum_stats_merge4
```

```python
#Generate a summary statistics table consisting of the mean, median, variance, standard deviation, and SEM of the tumor volume for each drug regimen.
# Using the aggregation method, produce the same summary statistics in a single line
drug_regimen = cleaned_combined_df.groupby(['Drug Regimen'])[['Tumor Volume (mm3)']].agg(['mean','median','var','std','sem'])
drug_regimen
```
